## JCPenney Data
There are 5 data files: 
`products.csv`, `users.csv`, `reviews.csv`,

Let's **import** the data and understand what information we have!

```python
# filepath to folder containing JCPenney data
directory = "C:/Users/Darragh/Desktop/RMData/jcpenney" 

# list of file names in JCPenney folder to load in
files = ['products.csv', 'users.csv', 'reviews.csv',
         'jcpenney_reviewers.json','jcpenney_products.json'
        ]
```

- Load in JCPenney data (using `read_csv()` and `read_json()`) as a dataframe (df)
- Then store in a dictonary for easy access. 

```python
# initialise dictionary
jcpenney_dfs = {} 

# for each file in jcpenney folder..
for file_name in files:  
    path = os.path.join(directory, file_name) # Store path to file 

    # Load .csv file as df and add to dictionary (key = file_name)
    if file_name.endswith('.csv'):      
         jcpenney_dfs[file_name] = pd.read_csv(path) 

    #^same as before but for .json files
    elif file_name.endswith('.json'):    
        jcpenney_dfs[file_name] = pd.read_json(path, lines = True)   
                #each line in .json file is a json object^^
        
# check files have been added correctly
print(jcpenney_dfs.keys())
```


####  We can access any file using the following:  `df = jcpenney_dfs[file_name]`

## Check total and average score reviews for all JC Penney products

```python
# get jcpenney_products.json df
df= jcpenney_dfs['jcpenney_products.json'].copy()

#show first 3 rows
display(df.head(3))
#display(df['Reviews'].iloc[1])
display(df.columns)
```

#### We see multiple rows for the same `product`, therefore we need to collate data so each product has it's own row.

#### We first filter for the columns needed:
- `sku` i.e. Stock Keeping Unit
-  `average_product_rating`
-  `total_number_reviews`.

```python
# filter columns
df = df[['sku','average_product_rating','total_number_reviews']]
df.head(4)
```

#### Next we group rows by `sku` and calculate the true `average_product_rating` and `total_number_reviews` for each product.

```python
df = df.groupby('sku').agg(                                  # for each sku (i.e. product)..
    average_product_rating = ('average_product_rating','mean') # calculate mean product rating
    total_number_reviews = ('total_number_reviews', 'sum')    #  calculate sum of reviews
)                                                          # return as new df
```

```python
df
```

#### `Reviews` column in `jcpenney_products.json` contains list of dictionaries, each dictionary contains info on user review (namely `User`, `Review` and `Score`)

#### We will use `explode()` to expand each review into a row

```python
df= df.explode('Reviews')
df.head(10)
#display(
```

```python
#from each review, will take score and store in new column
df['Score'] = df['Reviews'].str['Score']

#filter for sku and score
df = df[['sku','Score']]

#convert dtype sku -> string
df['sku'] = df['sku'].astype('string')

display(df.head(10))
```

## Jcpenney Discount

```python
#Load in jcpenney_products.json
df = jcpenney_dfs['jcpenney_products.json'].copy() #copy() to prevent warnings

display(df.head(10))

#'uniq_id',

# filter for columns
df = df[['sku','name_title', 'list_price', 'sale_price']]

# convert dtypes of columns
# SKU + Name -> string and list_price + sale_price -> float
df['sku'] = df['sku'].astype('string')
df['name_title'] = df['name_title'].astype('string')
df['list_price'] = pd.to_numeric(df['list_price'])
df['sale_price'] = pd.to_numeric(df['sale_price'], errors = 'coerce')
                                    #^any entry not able to be numeric will be NaN instead

df.info()
```

## Cleaning data

```python
# counter for rows in df
rows = len(df)
print(f'Total: {rows} rows in df.')

# filter out rows containing 'NaN' 
df.dropna(subset=['sku', 'name_title', 'list_price', 'sale_price'], inplace=True)

# print NaN rows removed
nan_count = rows - len(df)
rows = len(df)
print(f'{nan_count} rows contain "NaN" -> {rows} rows remaining.')



# filter out negative values (<0) in 'list_price' and 'sale_price' column
df = df[ (df['list_price'] > 0) & (df['sale_price'] > 0) ]

# print no. rows removed
negative_count = rows - len(df)
rows = len(df)
print(f'{negative_count} rows contain negative values -> {len(df)} rows remain.')


# filter out duplicate rows
df = df.drop_duplicates()

dup_count = rows - len(df)
print(f'{dup_count} rows contain duplicates -> {len(df)} rows remain.')

df.head(10)
```

## Discount % 
- Calculate %discount between list and sale price.
- Store under new column 'discount_percentage'

```python
#store numerical difference in column 'discount'
df['discount'] = df['list_price'] - df['sale_price'] 

# calculate %discount, store in new column
df['discount_percentage'] = ((df['discount'] / df['list_price']) * 100).round(2)
```

```python
# Query to find top 10 products by discount percentage
df.sort_values(by = 'discount_percentage', ascending = False).iloc[:10]

# General stats on discount percentage
display(df['discount_percentage'].describe().round(2))

#average discount = 41.5% acorss 4992 products
#
```

```python
# boxplot

sns.boxplot(y=df['discount_percentage'], color = 'skyblue')
plt.ylabel('Discount percentage (%)')
plt.show()
```

## Ratings Distributions by Age Category

```python
df = jcpenney_dfs['reviews.csv']

sns.countplot(x = 'Score',
              data = df, 
              palette ='rocket')
plt.show()
```

```python
#Load in jcpenney_products.json + users.csv into dfs
df = jcpenney_dfs['jcpenney_products.json']
df_user = jcpenney_dfs['users.csv']

#Review column in jcpenney_products.json contains list of dictonaries of Username, Review and Score
#Convert nested dictonary to df, save as df_score
df_score = pd.json_normalize(
    df['Reviews'].explode() )

#Convert to str
df_score['User'] = df_score['User'].astype('string')
df_score.head()
#df_user.head()
```

```python
# Convert user.csv column data types: Username -> str //   DOB -> datetime   // State -> category //
df_user['Username'] = df_user['Username'].astype('string')
df_user['DOB'] = pd.to_datetime(df_user['DOB'])
df_user['State'] = df_user['State'].astype('category')

#rename column username -> user to join df_score with df_user
df_user = df_user.rename(columns ={'Username':'User'})
```

```python
#create new age column from DOB
today = datetime.date.today()
df_user['Age'] = (
    (today - df_user['DOB'].dt.date) / pd.Timedelta(days=365.25)
    ).round().astype(int)                             #^account for leap years
                                                                       
print(df_score.head())
print(df_user.head())
```

```python
#Need to categorise age via generation

bins = [18,25,35,45,55,65,80]

age_cat = ['18-24',
    '25-34',
    '35-44',
    '45-54',
    '55-64',
    '65+']

#creating age category column from 'Age' column 
#right=false to ensure boundary ages e.g. 25 are include in correct 25-34 bin
df_user['Age_Category'] = pd.cut(df_user['Age'],  
                            bins=bins,
                            labels=age_cat, 
                            right=False) 

#ensure categories are ordered for sorting and plotting
df_user['Age_Category'] = df_user['Age_Category'].astype(
    pd.CategoricalDtype(categories=age_cat, ordered=True))

```

```python
# join df_user and df_score, new df called df_userscore
df_userscore = pd.merge(df_user, df_score, on='User', how = 'inner')


print(df_userscore.info(10))
display(df_userscore.head())
```

```python
df_userscore = df_userscore.drop_duplicates()
df_userscore.info()
```

```python
#Calculate average, total and std of review scores, by age category
score_by_age = df_userscore.groupby('Age_Category').agg(
    average = ('Score', 'mean'),
    total = ('Score', 'count'),
    std = ('Score','std')
).sort_values(by='Age_Category',ascending = False)

print(score_by_age)

##############################################################

#calculate total score given by age, by rating
rating_by = pd.crosstab(index=df_userscore['Age'],
                        columns=df_userscore['Score'])
print(rating_by)
```

```python
sns.boxplot(y=df_userscore['Score'],
            x=df_userscore['Age_Category'],
            color = 'skyblue')
plt.ylabel('Score (out of 5)')
plt.show()
```
