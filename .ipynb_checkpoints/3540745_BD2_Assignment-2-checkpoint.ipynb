## Check total and average score reviews for all JC Penney products

```python
# get jcpenney_products.json df
df= jcpenney_dfs['jcpenney_products.json'].copy()

#show first 3 rows
display(df.head(3))
```

#### We see multiple rows for the same `product`, therefore we need to collate data so each product has it's own row.

#### We first filter for the columns needed:
- `sku` i.e. Stock Keeping Unit
-  `average_product_rating`
-  `total_number_reviews`.


<br>

#### And clean up data (e.g. remove rows with empty rows, and keep plausible results (rating 1-5))

```python
# filter columns
df = df[['sku','average_product_rating','total_number_reviews']]
display(df.head(4))
display(df.shape)


# remove rows with empty sku entries
df = df[ df['sku'] != '']
display(df.shape)

#keep ratings 1-5 only
df = df[df['average_product_rating'].between(1,5)]
display(df.shape)
```

#### Next we group rows by `sku` and calculate the true `average_product_rating` and `total_number_reviews` for each product.

```python
#using .agg(), we create new columns and
# calculate mean product rating,
# calculate total number of reviews and
# return as new df, `df_sku_stats`
df_sku_stats = df.groupby('sku').agg(
    average_product_rating = ('average_product_rating','mean'), 
    total_number_reviews = ('total_number_reviews', 'sum'))    

display(df_sku_stats.head(3))

```

#### `df_sku_stats` contains all products with their respective `average_product_rating` and `total_number_review`.
<br>

####  Let's visualise their distributions for all products.


```python
p=3
df2 = df_sku_stats[df_sku_stats['total_number_reviews']>=p]
df2.head()
```

```python
#Histogram for average product rating
plt.figure(figsize = (6,4))

sns.histplot(x= df2['average_product_rating'],
        bins = 16,
        #kde = True,
        color = '#D20001',
        edgecolor='black')


plt.title('average product rating, all products, >3 reviews')
plt.xlabel('Average Product Rating')
plt.ylabel('Count')
plt.xticks([1, 2, 3, 4, 5])
plt.grid(linestyle = '--', alpha = 0.3)

plt.show()

```

```python
# Histogram for number of product reviews.

plt.figure(figsize = (6,4))

sns.histplot(x=df_sku_stats['total_number_reviews'],
        bins = 50,
        color = '#D20001',
        edgecolor = 'black')


plt.title('total_number_reviews top')
plt.xlabel('X-label')
#plt.axis([1, 150, 0, 2700])
plt.xticks(np.arange(0, 150, 10))

plt.yscale('log')
plt.ylabel('count (log)')

plt.grid(linestyle = '--', alpha = 0.4)


plt.show()
```

## Looking at brand insights

```python
#load in `jcpenney_products.json` df
df = jcpenney_dfs['jcpenney_products.json'].copy()

#filter columns
df = df[['sku','brand']]

#convert dtype sku and brand --> str
df['sku'] = df['sku'].astype('string')
df['brand'] = df['brand'].astype('string')

#convert brands to lowercase
#just in case merging of sku is stopped by capital letters
df['brand'] = df['brand'].str.lower()
display(df.head())
```

#### Clean up df by removing rows without `sku` + removing duplicate rows with same `sku` and `brand`

```python
#show starting no. rows in df
print(f'Starting rows = {len(df)}')

#drop empty sku rows
df = df[ df['sku'] != '']
print(f'Removing empty sku entries, Rows = {len(df)}')

#drop rows with same sku and same brand
df = df.drop_duplicates()
print(f'Removing duplicate entries, Rows = {len(df)}')
```

```python
# Check if the number of unique SKUs equals the total number of rows
# i.e. are all sku associated with only 1 brand?
sku_unique = df['sku'].nunique() == len(df)
print(sku_unique)
```

Since result is FALSE, we need to check which sku are associated with multiple brands and append appropriately 

```python
#check which skus have multiple brands associated

#group by unique skus and filter for sku's with >1 brand associated
find_sku = df.groupby('sku')['brand'].nunique()
dup_brand = find_sku[find_sku>1].index

#return df (dup_df) to check position of multi-brand associated skus
dup_df = df[ df['sku'].isin(dup_brand)].copy()
dup_df.head()
```

#### pp5002940877 + pp5005280084 are associated with multiple brands, however seems to be different wordings for a similar brand
#### Lets check which brand wording appears most frequently and we'll remove rows with least frequently occurring

```python
#Finding freq of all brands 
all_brands = df['brand'].value_counts()

# get brands which appear with multiple sku (from previous)
# and find how often they occur 
conflict_brands = dup_df['brand'].unique()
freq = all_brands.loc[conflict_brands]

print(freq.sort_values(ascending=False))
```

Appears `studio`, `jcp home collection` are the least frequent..

```python
#Let's remove entries by their index (7731 + 7739)
print(len(df))
df.drop(labels=[7731,7739], axis=0, inplace=True)
print(len(df))
```

```python
# Again, check if unique SKUs are associate with 1 brand only
print(sku_unique)
```

Now that the result is TRUE, we can continue with the data now each `sku` is associated with a `brand`

---------------------------------------------------------------
#### Merge dfs `df_sku_stats` with `df_brands` to get stats about each brand

```python
#Set sku as index and save df as `df_brands` for later
df_brands = df.set_index('sku')
df_brands.head(3)
```

```python
#Check number of rows in each df
display(df_sku_stats.head(2))
print(f'df_sku_stats rows: {len(df_sku_stats)}')

display(df_brands.head(2))
print(f'df_brands rows: {len(df_brands)}')
```

##### Both df have same number of rows, let's merge them together!

```python
#Merge df using Inner join on the sku
df = pd.merge(df_sku_stats, df_brands, on='sku', how = 'inner')
```

```python
display(df.head(2))
print(f'\nNumber of rows: {len(df)}')
```

```python
#reset index so sku is a normal column
df = df.reset_index()
```

#### Next we group data by brand to get insights on individual brand performance

```python
#Using .agg() we can get brand average rating, total ratings 
#and number of products used in aggregation
#Then save df as `brand_stats_df` for future recall
brand_stats_df = df.groupby('brand').agg(
    average_rating=('average_product_rating', 'mean'),
    total_reviews=('total_number_reviews', 'sum'),
    number_of_products=('sku', 'count')
)

print(f'{brand_stats_df.head()}  \n\nNumber of brands: {len(brand_stats_df)}')
```

## Jcpenney Discount

```python
#Load in jcpenney_products.json
df = jcpenney_dfs['jcpenney_products.json'].copy() #copy() to prevent warnings
display(df.head(1))

# filter for columns
df = df[['sku','name_title', 'list_price', 'sale_price']]

# convert dtypes of columns
# SKU + Name -> string and list_price + sale_price -> float
df['sku'] = df['sku'].astype('string')
df['name_title'] = df['name_title'].astype('string')
df['list_price'] = pd.to_numeric(df['list_price'], errors = 'coerce')
df['sale_price'] = pd.to_numeric(df['sale_price'], errors = 'coerce')
                                    #^any entry not able to be numeric will be NaN instead

df.info()
```

#### Looking at the data, there seems to be alot of NaN values in the `list_price` column, of 7982 entries only 5816 are non-null
#### We will assume that `list_price` = `sale_price` (i.e. no discount)

```python
# fill list_price with sale_price if sale_price is non-null
df['list_price'] = df['list_price'].fillna(df['sale_price'])
```

## Cleaning data

```python
# check no. rows in df
print(f'Total: {len(df)} rows in df.')

# filter out rows with empty entries for sku
df = df[ df['sku'] != '']
print(f'Remove rows contain empty sku -> {len(df)} rows remaining.')

# filter out rows containing 'NaN' 
df.dropna(subset=['sku', 'name_title', 'list_price', 'sale_price'], inplace=True)
print(f'Remove rows contain "NaN" -> {len(df)} rows remaining.')

# filter out negative values (<0) in 'list_price' and 'sale_price' column
df = df[ (df['list_price'] > 0) & (df['sale_price'] > 0) ]
print(f'Remove rows contain negative values -> {len(df)} rows remain.')

# filter out duplicate rows
df = df.drop_duplicates()
print(f'Remove rows contain duplicates -> {len(df)} rows remain.')

df.head(5)
```

```python
display(df['sku'].value_counts())

#pp5006400007
#pp5003910685
#pp5005921642 
```

## Discount % 
- Calculate %discount between list and sale price.
- Store under new column 'discount_percentage'

```python
#store numerical difference in column 'discount'
df['discount'] = df['list_price'] - df['sale_price'] 

# calculate %discount, store in new column
df['discount_percentage'] = ((df['discount'] / df['list_price']) * 100).round(2)

df_discount = df.copy()
```

## Ratings Distributions by Age Category

```python
df = jcpenney_dfs['reviews.csv']

sns.countplot(x = 'Score',
              data = df, 
              palette ='rocket')
plt.show()
```

```python
#Load in jcpenney_products.json + users.csv into dfs
df = jcpenney_dfs['jcpenney_products.json']
df_user = jcpenney_dfs['users.csv']

#Review column in jcpenney_products.json contains list of dictonaries of Username, Review and Score
#Convert nested dictonary to df, save as df_score
df_score = pd.json_normalize(
    df['Reviews'].explode() )

#Convert to str
df_score['User'] = df_score['User'].astype('string')
df_score.head()
```

```python
# Convert user.csv column data types: Username -> str //   DOB -> datetime   // State -> category //
df_user['Username'] = df_user['Username'].astype('string')
df_user['DOB'] = pd.to_datetime(df_user['DOB'])
df_user['State'] = df_user['State'].astype('category')

#rename column username -> user to join df_score with df_user
df_user = df_user.rename(columns ={'Username':'User'})
```

```python
#create new age column from DOB
today = datetime.date.today()
df_user['Age'] = ((today - df_user['DOB'].dt.date) / pd.Timedelta(days=365.25)
    ).round().astype(int)                             #^account for leap years
                                                                       
display(df_score.head(3))
display(df_user.head(3))
```

```python
#Need to categorise age via generation

bins = [18,25,35,45,55,65,80]

age_cat = ['18-24','25-34','35-44','45-54','55-64','65+']

#creating age category column from 'Age' column 
#right=false to ensure boundary ages e.g. 25 are include in correct 25-34 bin
df_user['Age_Category'] = pd.cut(df_user['Age'],  
                            bins=bins,
                            labels=age_cat, 
                            right=False) 

#ensure categories are ordered for sorting and plotting
df_user['Age_Category'] = df_user['Age_Category'].astype(
    pd.CategoricalDtype(categories=age_cat, ordered=True))

```

```python
# join df_user and df_score, new df called df_userscore
df_userscore = pd.merge(df_user, df_score, on='User', how = 'inner')


print(df_userscore.info(10))
display(df_userscore.head())
```

```python
df_userscore = df_userscore.drop_duplicates()
df_userscore.info()
```

```python
#Calculate average, total and std of review scores, by age category
score_by_age = df_userscore.groupby('Age_Category').agg(
    average = ('Score', 'mean'),
    total = ('Score', 'count'),
    std = ('Score','std')
).sort_values(by='Age_Category',ascending = False)

display(score_by_age)

##############################################################

#calculate total score given by age, by rating
rating_by = pd.crosstab(index=df_userscore['Age_Category'],
                        columns=df_userscore['Score'])
display(rating_by)
```
